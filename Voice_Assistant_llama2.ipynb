```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch

# Load Llama-2-7B Model
model_name = "shreeyans1298/Llama-2-7b-chat-finetune-chatbot"

tokenizer = AutoTokenizer.from_pretrained(model_name)

# Handle GPU or CPU automatically
device = "cuda" if torch.cuda.is_available() else "cpu"
dtype = torch.float16 if torch.cuda.is_available() else torch.float32

model = AutoModelForCausalLM.from_pretrained(
    model_name,
    torch_dtype=dtype,
    device_map="auto"
).to(device)

def get_response(user_input):
    # Tokenize input and move to the correct device
    inputs = tokenizer(user_input, return_tensors="pt").to(device)

    # Generate response with controlled randomness
    outputs = model.generate(
        **inputs,
        max_length=250,   # Adjust max response length
        do_sample=True,   # Enable randomness
        temperature=0.7,  # Control creativity (0.7 is balanced)
        top_p=0.9,        # Nucleus sampling for diverse answers
        repetition_penalty=1.2  # Reduce repeating responses
    )

    response_text = tokenizer.decode(outputs[0], skip_special_tokens=True)

    print(f"Raw Model Output: {response_text}")  # Debugging output

    return response_text
```

```python
import whisper

whisper_model = whisper.load_model("base")

def transcribe_audio(audio_path):
    result = whisper_model.transcribe(audio_path)
    return result["text"]
```

```python
from gtts import gTTS

def text_to_speech(text, filename="output.mp3"):
    tts = gTTS(text)
    tts.save(filename)
    return filename
```

```python
import gradio as gr

import os
import traceback

def chat_with_bot(audio):
    try:
        # Step 1: Debug Audio File
        print(f"Raw Audio Input: {audio}")

        # Step 2: Ensure audio is valid
        if not audio:
            print("Error: No audio file received!")
            return "Error: No audio file received!", None

        # Step 3: Get the correct file path
        audio_path = audio.name if hasattr(audio, "name") else audio
        print(f"Received Audio File: {audio_path}")

        # Ensure file exists
        if not os.path.exists(audio_path):
            print("Error: Audio file not found!")
            return "Error: Audio file not found!", None

        # Step 4: Transcribe audio
        user_input = transcribe_audio(audio_path)
        print(f"User said: {user_input}")

        # Step 5: Generate chatbot response
        bot_reply = get_response(user_input)
        print(f"Chatbot Reply: {bot_reply}")

        # Step 6: Convert to speech
        audio_output = text_to_speech(bot_reply)
        print(f"Generated Voice File: {audio_output}")

        return bot_reply, audio_output

    except Exception as e:
        print("Full Error Traceback:")
        print(traceback.format_exc())  # Print full error details
        return f"Error: {str(e)}", None



iface = gr.Interface(
    fn=chat_with_bot,
    inputs=gr.Audio(sources=["microphone"], type="filepath", label="Speak to the bot"),
    outputs=[
        gr.Textbox(label="Bot Reply"),
        gr.Audio(label="Bot Voice Reply")
    ],
    title="Voice Assistant Chatbot Which explain your words"
)
iface.launch()
```
